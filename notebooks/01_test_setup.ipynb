```python
mongodb_url = os.getenv("MONGODB_URL")
client = MongoClient(mongodb_url)

# Tester la connexion
try:
    client.admin.command('ping')
    print("‚úÖ Connexion MongoDB r√©ussie !")
    
    # Cr√©er une collection de test
    db = client["real_estate"]
    collection = db["test"]
    
    # Ins√©rer un document test
    test_doc = {"test": "Hello from Jupyter!", "value": 42}
    result = collection.insert_one(test_doc)
    print(f"üìù Document ins√©r√© avec l'ID : {result.inserted_id}")
    
    # Lire le document
    retrieved = collection.find_one({"test": "Hello from Jupyter!"})
    print(f"üìñ Document lu : {retrieved}")
    
    # Nettoyer (supprimer le test)
    collection.delete_one({"_id": result.inserted_id})
    print("üóëÔ∏è Document test supprim√©")
    
except Exception as e:
    print(f"‚ùå Erreur : {e}")
finally:
    client.close()
```
